# Which band to choose for point-like sources detection ?

The goal of this short notebook is to answer this question by looking at the different check images produced by the [point-like sources detection program](02_SExtractor.ipynb) from this project and manually select the one giving the best result. From my analysis, I chose the JWST F200W band as it gave the clearest star line, while also being stable enough between different fields to ensure that the set threshold for the linear regression should work for most cases.

```python
import numpy as np
import matplotlib.pyplot as plt
from dja_sepp.utils import color_dict
```

```python
filter_list = ['f090w', 'f115w', 'f150w', 'f200w', 'f277w', 'f356w', 'f444w']

fig, axs = plt.subplots(2,4, figsize=(12,5), sharex=True, sharey=True, gridspec_kw = {'wspace':0, 'hspace':0})
for i, filter in enumerate(filter_list):
    ax = axs.flatten()[i]
    img = plt.imread(f"/home/ec2-user/DAWN/DJA-SEpp/checkimages/starLine_gds-grizli-v7.0-{filter}-clear_drc_sci.png")
    ax.imshow(img)
    ax.axis('off')
    ax.text(0.68, 0.83, filter.upper(), transform=ax.transAxes, fontsize='x-large', fontweight='bold', color=color_dict[filter.upper()])

for j in range(len(filter_list), len(axs.flatten())):
    fig.delaxes(axs.flatten()[j])
    
fig.tight_layout()
plt.show()
```
